## 1. Take integer and print its square

```java
import java.util.Scanner;

class Square {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        System.out.println(n * n);
    }
}
```

## 2. Take two numbers and swap them

### a) With third variable

```java
import java.util.Scanner;

class SwapWithTemp {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        int a = sc.nextInt();
        int b = sc.nextInt();

        int temp = a;
        a = b;
        b = temp;

        System.out.println(a + " " + b);
    }
}
```

### b) Without third variable

```java
import java.util.Scanner;

class SwapWithoutTemp {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        int a = sc.nextInt();
        int b = sc.nextInt();

        a = a + b;
        b = a - b;
        a = a - b;

        System.out.println(a + " " + b);
    }
}
```

Interview Note:
This method may cause **overflow**, so temp-variable method is safer.

## 3. Find minimum of three numbers

```java
import java.util.Scanner;

class MinOfThree {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        int a = sc.nextInt();
        int b = sc.nextInt();
        int c = sc.nextInt();

        int min = a;
        if (b < min) min = b;
        if (c < min) min = c;

        System.out.println(min);
    }
}
```

## 4. Print ASCII value of a character

```java
import java.util.Scanner;

class AsciiValue {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        char ch = sc.next().charAt(0);

        int ascii = ch;
        System.out.println(ascii);
    }
}
```

## 5. Demonstrate buffer issue using Scanner

### Problem Code

```java
import java.util.Scanner;

class BufferIssue {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);

        int x = sc.nextInt();
        String s = sc.nextLine();

        System.out.println(x);
        System.out.println(s);
    }
}
```

Issue:

* `nextInt()` leaves a newline in buffer
* `nextLine()` reads that newline

### Correct Code (Fix)

```java
import java.util.Scanner;

class BufferFix {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);

        int x = sc.nextInt();
        sc.nextLine(); // clears buffer
        String s = sc.nextLine();

        System.out.println(x);
        System.out.println(s);
    }
}
```

Interview Line:

* Buffer issue happens when mixing token-based and line-based input methods.

# Swapping Two Numbers – Methods and Problems

Assume:

```java
int a = 5;
int b = 3;
```

---

## 1. Using Third Variable (Recommended)

```java
int temp = a;
a = b;
b = temp;
```

### Advantages

* Simple
* Readable
* No risk
* Interview-safe

### Problems

* Uses extra variable (negligible in real life)

Interview Verdict:
Best and safest method.

---

## 2. Using Addition and Subtraction

```java
a = a + b;
b = a - b;
a = a - b;
```

### Advantages

* No extra variable

### Problems

* **Integer overflow**
* Fails for large values
* Less readable

Example:

```java
int a = Integer.MAX_VALUE;
int b = 1;
```

Interview Verdict:
Avoid in real applications.

---

## 3. Using Multiplication and Division

```java
a = a * b;
b = a / b;
a = a / b;
```

### Advantages

* No extra variable

### Problems

* **Division by zero** if `a` or `b` is 0
* Overflow risk
* Slower than addition

Interview Verdict:
Unsafe and not recommended.

---

## 4. Using XOR (Bitwise Swap)

```java
a = a ^ b;
b = a ^ b;
a = a ^ b;
```

### Advantages

* No extra variable
* No overflow
* Fast

### Problems

* Hard to read
* Confusing in interviews
* Fails when `a` and `b` refer to same variable

Example problem:

```java
a = a ^ a; // result becomes 0
```

Interview Verdict:
Good for concept discussion, not for production.

---

## 5. Using Single-Line Java Trick (Not Recommended)

```java
b = a + b - (a = b);
```

### Problems

* Very confusing
* Poor readability
* Bad practice

Interview Verdict:
Never use.

---

## 6. Swapping Using Method (Important Java Concept)

```java
static void swap(int a, int b) {
    int temp = a;
    a = b;
    b = temp;
}
```

### Problem

* **Does NOT work in Java**

Reason:

* Java is **pass by value**
* Only copies are swapped

Interview Trap Question:
"Why swap method does not work in Java?"

Correct Answer:

* Because Java passes a copy of primitive values.

---

## 7. Swapping Using Objects (Works)

```java
class Pair {
    int a, b;
}

static void swap(Pair p) {
    int temp = p.a;
    p.a = p.b;
    p.b = temp;
}
```

### Why This Works

* Reference is passed by value
* Object data is modified

Interview Importance:
Very high.
