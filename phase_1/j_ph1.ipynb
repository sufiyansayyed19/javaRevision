## 2. Keywords and Identifiers

### Keywords

* Reserved words with predefined meaning
* Examples: `class`, `public`, `static`, `void`, `int`

Rules:

* Cannot be used as variable or class names
* All are lowercase

### Identifiers

* Must not start with digit
* Cannot contain spaces
* Cannot be keywords

## 2. Take two numbers and swap them

### a) With third variable

```java
import java.util.Scanner;

class SwapWithTemp {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        int a = sc.nextInt();
        int b = sc.nextInt();

        int temp = a;
        a = b;
        b = temp;

        System.out.println(a + " " + b);
    }
}
```

### b) Without third variable

```java
import java.util.Scanner;

class SwapWithoutTemp {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        int a = sc.nextInt();
        int b = sc.nextInt();

        a = a + b;
        b = a - b;
        a = a - b;

        System.out.println(a + " " + b);
    }
}
```

Interview Note:
This method may cause **overflow**, so temp-variable method is safer.

## 3. Find minimum of three numbers

```java
import java.util.Scanner;

class MinOfThree {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        int a = sc.nextInt();
        int b = sc.nextInt();
        int c = sc.nextInt();

        int min = a;
        if (b < min) min = b;
        if (c < min) min = c;

        System.out.println(min);
    }
}
```

## 4. Print ASCII value of a character

```java
import java.util.Scanner;

class AsciiValue {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        char ch = sc.next().charAt(0);

        int ascii = ch;
        System.out.println(ascii);
    }
}
```

## 5. Demonstrate buffer issue using Scanner

### Problem Code

```java
import java.util.Scanner;

class BufferIssue {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);

        int x = sc.nextInt();
        String s = sc.nextLine();

        System.out.println(x);
        System.out.println(s);
    }
}
```

Issue:

* `nextInt()` leaves a newline in buffer
* `nextLine()` reads that newline

### Correct Code (Fix)

```java
import java.util.Scanner;

class BufferFix {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);

        int x = sc.nextInt();
        sc.nextLine(); // clears buffer
        String s = sc.nextLine();

        System.out.println(x);
        System.out.println(s);
    }
}
```

Interview Line:

* Buffer issue happens when mixing token-based and line-based input methods.
